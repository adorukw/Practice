## VS Code中使用Jupyter Notebook指南

#### 1. 快捷键

- 运行当前单元格：`Ctrl + Enter`
- 运行当前单元格并选择下一个：`Shift + Enter`
- 运行所有单元格：`Alt + Enter`
- 切换到上一个单元格：`Ctrl + Alt + ↑`
- 切换到下一个单元格：`Ctrl + Alt + ↓`
- 插入单元格：`Ctrl + Shift + -`
- 格式化代码：`Ctrl + Shift + I`

```python
# %% [markdown]
# ## 1. 变量作用域示例
# 在 Jupyter 中，变量是**全局的**

# %%
# 单元格1：定义变量
x = 10
y = "hello"
my_list = [1, 2, 3]

print(f"单元格1: x = {x}, y = {y}, my_list = {my_list}")
# 运行后，这些变量在整个notebook中都可访问

# %%
# 单元格2：访问单元格1的变量
print(f"单元格2: 我可以访问 x = {x}")
print(f"单元格2: 我可以访问 y = {y}")
print(f"单元格2: 我可以修改 my_list: {my_list}")

# 修改单元格1中定义的变量
my_list.append(4)
print(f"单元格2: 修改后的 my_list = {my_list}")

# %%
# 单元格3：查看修改后的变量
print(f"单元格3: 看到my_list已被修改: {my_list}")
print(f"单元格3: x仍然是: {x}")

# 再修改变量
x = 20
print(f"单元格3: 修改x为: {x}")
```
