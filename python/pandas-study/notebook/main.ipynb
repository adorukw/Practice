```python
import pandas as pd

print(pd.__version__)

data = {
    '姓名': ['小明', '小红', '小刚'],
    '年龄': [18, 20, 19],
    '城市': ['北京', '上海', '广州']
}

df=pd.DataFrame(data)
print("从字典创建的DataFrame：\n")
print(df)
print("数据类型：",type(df))

data_list=[
    ['小明', 18, '北京'],
    ['小红', 20, '上海'],
    ['小刚', 19, '广州']
]

df2=pd.DataFrame(data_list,columns=['姓名','年龄','城市'])
print("从列表创建的DataFrame：")
print(df2)
print("df和df2是否相等：",df.equals(df2))
```
